```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_redundant=0)
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
# Using linear kernel
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print("Classification_report",classification_report(y_test,y_pred))
print("Confusion matrix",confusion_matrix(y_test,y_pred))
print("Accuracy_score",accuracy_score(y_test,y_pred)*100)

```

```python
#using rbf fucntion
rbf=SVC(kernel='rbf')
rbf.fit(x_train,y_train)
y1_pred=rbf.predict(x_test)
print("Classification report:\n",classification_report(y_test,y1_pred))
print("Confusion_matrix:\n",confusion_matrix(y_test,y1_pred))
print("Accuracy_score",accuracy_score(y_test,y1_pred)*100)
```

```python
# using polynomial function
pol=SVC(kernel='poly')
pol.fit(x_train,y_train)
y2_pred=pol.predict(x_test)
print("Classification report:\n",classification_report(y_test,y2_pred))
print("Confusion_matrix:\n",confusion_matrix(y_test,y2_pred))
print("Accuracy_score",accuracy_score(y_test,y2_pred)*100)
```

```python
#using sigmoid function
sig=SVC(kernel='sigmoid')
sig.fit(x_train,y_train)
y3_pred=sig.predict(x_test)
print("Classification report:\n",classification_report(y_test,y3_pred))
print("Confusion matrix:\n",confusion_matrix(y_test,y3_pred))
print("Accuracy_score:",accuracy_score(y3_pred,y_test)*100)
```

# HyperParameter tuning with svc


```python
from sklearn.model_selection import GridSearchCV
# defining parameters range
param_grid={'C':[0.1,1,10,100,1000],
'gamma':[1,0.1,0.01,0.001,0.0001],
'kernel':['rbf']}
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
grid.fit(x_train,y_train)
grid.best_params_
```

```python
y4_pred=grid.predict(x_test)
print("classification_report:\n",classification_report(y4_pred,y_test))
print("confusion_matrix:\n",confusion_matrix(y4_pred,y_test))
print("Accuracy_score:",accuracy_score(y_test,y4_pred)*100)
```
